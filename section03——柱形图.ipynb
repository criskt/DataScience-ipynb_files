# 为什么销售会减少？


一款叫做《黑猫拼图》的社交游戏本月的销售额相较于上月有所下滑，于是想调查下滑的原因，并提升销售额。

### 【知识点】

- missingno 缺失值可视化
- pd.merge()  df.groupby()
- pd.to_datetime()
- plot/sharey

## 0、导入基本库

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import missingno as msno
import seaborn as sns
%matplotlib inline
```

## 1、读取数据

```python
dau_data = 'data/section3-dau.csv'
dpu_data = 'data/section3-dpu.csv'
install_data = 'data/section3-install.csv'
```

```python
dau = pd.read_csv(dau_data)
dpu = pd.read_csv(dpu_data)
ins = pd.read_csv(install_data)
```

```python
print(dau.info())
dau.head()
```

```python
print(dpu.info())
dpu.head()
```

```python
print(ins.info())
ins.head()
```

## 2、合并 DAU  INSTALL 的数据

```python
dapu = pd.merge(dau,ins,on=['user_id','app_name'])
```

```python
print(dapu.info())
dapu.head()
```

## 3、合并上述数据和 DPU 数据

```python
apin = pd.merge(dapu,dpu,on=['log_date','app_name','user_id'],how='left')
```

```python
print(apin.info())
apin.head()
```

观察数据完整性

```python
msno.matrix(apin)
```

### 将未消费用户的消费额设置为零

```python
apin.fillna(0,inplace=True)
```

```python
apin.info()
```

## 4、按月统计，区分新用户以及老用户

```python
# 转化为日期格式

apin['log_date'] = pd.to_datetime(apin['log_date'])
apin['install_date'] = pd.to_datetime(apin['install_date'])
```

```python
apin.head()
```

```python
# 新增加一列表示月份，一列表示新用户标签

apin['log_month'] = apin.log_date.dt.month
apin['ins_month'] = apin.install_date.dt.month
apin['is_new'] = apin.log_month == apin.ins_month
```

```python
apin['is_new'] = apin['is_new'].apply(lambda x : 'yes' if x else 'no')
```

```python
df = apin.groupby(['log_month','is_new'])['payment'].sum()
```

```python
df = df.reset_index()
```

```python
df
```

## 5、数据可视化

```python
fig,ax = plt.subplots(1,1,figsize=(12,8))

sns.barplot(x = df.log_month,y=df[df.is_new == 'no'].loc[:,'payment'],data=df,ax=ax,color='gray',label='no new')
sns.barplot(x = df.log_month,y=df[df.is_new == 'yes']['payment'],data=df,ax=ax,color='lightgray',
            bottom=df[df.is_new == 'no'].loc[:,'payment'],label='yes new')
plt.legend(loc='upper center',ncol=2,fontsize=20)
plt.ylim(0,300000)
sns.despine()
sns.set_context("poster", rc={"lines.linewidth": 10})
# sns.set_context("paper", font_scale=2.4)
# sns.set_style("ticks", {"xtick.major.size": 2, "ytick.major.size": 5})
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,8),sharey=True)
data = apin[apin.is_new == 'yes'][apin.payment >0]
axes[0].hist(data[data['log_month']==6]['payment'],bins=20,data=data,color='r',label='6yue')
axes[1].hist(data[data['log_month']==7]['payment'],bins=20,data=data,color='g',label='7yue')
axes[0].legend()
axes[1].legend()
# axes[1].ylim(0,70)
```
