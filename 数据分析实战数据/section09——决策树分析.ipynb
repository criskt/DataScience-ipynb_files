## 本篇旨在探讨用户的哪些行为会促使其成为长期用户 ###
·
·
·
* ### 一、问题引出 ### 
运营中，刚开始注册的用户往往会在不久后大量流失。而探究用户流失的原因并不可取，因为大量的轻度玩家属于自然流失，这样的探究是没有太大意义的。因此，如果想要留住用户，最好从资深玩家入手，了解他们从注册到成为重度玩家的行为特征，以优化运营。


* ### 二、特征分析 ###
游戏玩家的行为特征，也即玩家从游戏中能够体会到的‘乐趣’。数据分析中需要将这样的模糊性名词量化，这里将‘乐趣’分解为三类：
    * 战斗：对战其他用户
    * 协作：和其他用户协作共同打败敌方首领
    * 发送消息：向其他用户发送消息
    
    
* ### 三、数据分析方法 ###
    * 量化稳定来访 —— `N 日登录密度 = N 日内用户到访的天数 / N`。 此处将用户开始游戏后1周内的行为作为自变量，下一周的登录密度作为因变量进行分析）
    * 分析方法 —— `决策树`。 用决策树找出第一个以及第二、三个分裂属性，第一个分裂属性即代表对结果（稳定来访）影响最大的属性。
    * 分析属性 —— `实数、比率以及主成分`。 实数，即某社交行为发生的次数以及这些行为发生在多少天后；比率，即每天的访问次数占7日内的访问总次数的比率；主成分，即社交行为特征的主成分（当各种社交行为相互影响，或者没有什么行为的用户较多的情况下，使用独立的主成分能得到更好的分离）。
    * 将`类`作为自变量来使用 —— 太多的自变量`（（战斗、协作、消息） * （时间、次数） * （实数、比率、主成分））`放在一起分析，得出的结果很可能难以解释。因此，需要首先弄清楚用哪个分析属性分析那种社交行为最能说明稳定到访率，并在这一过程中确认哪种行为模式容易促使用户稳定来访。因此，需要将社交行为和分析属性组合成一个新的自变量（即：社交行为=[战斗、协作、消息]、分析属性=[实数、比率、主成分]）。使用这个新的自变量进行决策树分析，就能找出对稳定到访率影响最大的社交行为和分析属性的组合。组合成一个新变量的方法很多，本例将有类似行为的用户归到一起，使用kmeans聚类的方法。
    
* ### 四、大致流程 ###
    * 数据收集 — — 
        * install —— 新用户
        * dau —— 登录信息数据
        * battle —— 战斗（行为日志）
        * message —— 消息（行为日志）
        * help —— 协作（行为日志）
    * 数据加工 — —
        * 计算登录密度 — —
            * 将DAU和首次访问时间数据合并
            * 取得用户首次访问后第7-13天的数据
            * 计算出每个用户的登录密度
            * 每个分析对象用户与其登录密度合并
        * 生成用户首次访问后7天内的各个社交行为的数据 — —
            * 将各个社交行为的数据和用户首次访问时间的数据合并
            * 将时间限定在用户首次访问后的第0-6天
            * 将第 N 天的数据整理到同一列并按天排列
        * 聚类分析 — —
            * 通过3-6个类对各种行为 X 分析属性进行聚类
            * 将“行为名称_类的个数_分析属性”作为自变量
            * 合并登录密度与自变量
    * 数据分析 — —
        * 决策树分析 — —
            * 决策树最顶端节点即是影响最大的组合
            * 还可以观察叶节点，找到影响最大的一个节点
        * 探究每个协作主成分类的倾向 — — （书里得到的是协作主成分类）
            * 画出每个类的平均登录密度
            * 各个类中协作行为的比率随经过天数的变化
    * 结论 — —
        * `*协作行为*`逐渐增多的用户更容易稳定访问游戏！

```python
def read_bath(path):
    df = pd.DataFrame()
    for root, dirs, files in os.walk(path):
        for file in files:
            rfile = os.path.join(root,file)
            if rfile.split('.')[-1] == 'csv':
                rdf = pd.read_csv(rfile)
                df = df.append(rdf)
    return df
```

## 2、数据处理

### 2.1、计算登录密度


此处不需要用left方式合并

```python
# 合并 dau install 

dau_inst = pd.merge(dau, install, on='user_id', suffixes=('', '_inst'))   # how='left',不需要
print(dau_inst.shape)
dau_inst.head()
```

```python
# 限定为用户开始游戏后第7-13天的数据

for i in ['log_date','log_date_inst']:
    dau_inst[i] = pd.to_datetime(dau_inst[i])

dau_inst['elapsed_days'] = (dau_inst['log_date'] - dau_inst['log_date_inst']).dt.days

dau_inst_713 = dau_inst[(dau_inst['elapsed_days'] >= 7) & (dau_inst['elapsed_days'] <= 13)]
dau_inst_713.head()
```

```python
# 计算登陆密度

dau_inst_713_logds = dau_inst_713.groupby(['user_id'])['log_date'].count().reset_index()
dau_inst_713_logds['density'] = dau_inst_713_logds['log_date']/7
dau_inst_713_logds.drop(columns='log_date', inplace=True)
dau_inst_713_logds.head()
```

### 2.2、合并分析对象用户和登陆密度数据

```python
# 分析对象数据限制在 2013-06-01 和 2013-08-25 之内

target_install = install[(install['log_date']>='2013-06-01') & (install['log_date']<='2013-08-25')]
target_install.head()
```

```python
# 合并对象用户和登录密度数据

target_install_logds = pd.merge(target_install, dau_inst_713_logds, on='user_id', how='left')
target_install_logds.density.fillna(0, inplace=True)
target_install_logds.head()
```

### 2.3、生成关于战斗的数据

#### 2.3.1、生成实数数据

```python
#合并战斗行为数据和用户首次访问的数据

battle_inst = pd.merge(battle, install, on='user_id', suffixes=('','_inst'))    # , how='left'
battle_inst.head()
```

```python
# 计算用户进行战斗的时间和首次访问时间的时间差

for i in ['log_date','log_date_inst']:
    battle_inst[i] = pd.to_datetime(battle_inst[i])
battle_inst['elapsed_days'] = (battle_inst['log_date']-battle_inst['log_date_inst']).dt.days

# 删除杂质数据  也可以在下一步过滤
# battle_inst = battle_inst[battle_inst['elapsed_days']>=0].drop(columns='battle_count')
battle_inst.head()
```

```python
# 限定为用户进行战斗的时间和首次访问时间的时间差在1周以内的数据

battle_inst2 = battle_inst[(battle_inst['elapsed_days'] >= 0) & (battle_inst['elapsed_days'] <= 6)]
battle_inst2.head()
```

```python
# 将首次访问之后经过的天数按列排序

battle_inst2_cast = pd.pivot_table(battle_inst2, index='user_id', columns='elapsed_days', values='count', aggfunc='sum')
battle_inst2_cast.fillna(0, inplace=True)
battle_inst2_cast.columns = ['d0','d1','d2','d3','d4','d5','d6']
battle_inst2_cast.head(8)
```

#### 2.3.2、生成比率数据

```python
# 生成比率数据

battle_inst2_cast_prop = battle_inst2_cast.T/battle_inst2_cast.sum(axis=1)
battle_inst2_cast_prop = battle_inst2_cast_prop.T
battle_inst2_cast_prop.head(8)
```

#### 2.3.3、生成主成分（PCA）数据

```python
# PCA

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

# 主成分分析前先进行标准化
bicc = scale(battle_inst2_cast)
pca = PCA(whiten=False, svd_solver='auto')

b_pca = pca.fit_transform(bicc)

b_pca = pd.DataFrame(b_pca)
b_pca.columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7']
b_pca.index = battle_inst2_cast.index
print(pca.explained_variance_ratio_)
b_pca.head()
```

### 2.6、对行为日志进行剧聚类

```python
# 用来生成关于类数据的函数

from sklearn.cluster import KMeans

def clus(data,cols):
    df = pd.DataFrame()
    for a,b in zip(data,cols):
        coo = []
        labs = []
        SSE = []
        dff = pd.DataFrame()
        for i in range(3,7):
            km = KMeans(n_clusters=i+1).fit(a)
            lab = km.labels_
            labs.append(lab)
            col = b+'_'+str(i)
            coo.append(col)
        dff = pd.DataFrame(labs,index=coo,columns=a.index).T
        if len(df.columns)!=0:
            df = pd.concat([df,dff],axis=1)
        else:
            df = dff
    return df.reset_index()
```

```python
# 战斗

data1 = [b_pca, battle_inst2_cast, battle_inst2_cast_prop]
cols1 = ['b_pca','battle_inst2_cast','battle_inst2_cast_prop']

battle_cluster = clus(data1,cols1)
battle_cluster.head()
```

```python
# 发送消息

data2 = [m_pca, msg_inst2_cast, msg_inst2_cast_prop]
cols2 = ['m_pca','msg_inst2_cast','msg_inst2_cast_prop']

msg_cluster = clus(data2,cols2)
msg_cluster.head()
```

```python
# 协作

data3 = [h_pca, hlp_inst2_cast, hlp_inst2_cast_prop]
cols3 = ['h_pca','hlp_inst2_cast','hlp_inst2_cast_prop']

hlp_cluster = clus(data3,cols3)
hlp_cluster.head()
```

## 3、决策树分析的执行和可视化

```python
# 产生数据  

y = cluster_data2['density'].reshape(-1,1)

# y = yf


x = cluster_data2.drop(columns=['user_id','density'])

x.head()
```

```python
# 网格搜索最佳参数

from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0, train_size=0.8)
# columns = X_train.columns


# estimator = DecisionTreeClassifier( criterion='entropy' )
estimator = DecisionTreeRegressor( criterion='mse' )

param_ = {'max_features':range(3,9,1),'min_samples_leaf':range(2,10,1),'min_samples_split':range(2,9,1),'max_depth':range(2,8,1)}

grid = GridSearchCV(estimator, param_grid = param_, cv=5)
grid.fit(x,y)

# gsearch1.grid_scores_, 
# grid.best_params_, 
grid.best_score_
```

```python
# 使用最佳参数训练

# tc = DecisionTreeRegressor(criterion="mse",max_depth=4, max_features=6, min_samples_split=5, min_samples_leaf=5)   # 
# tc = DecisionTreeClassifier(criterion='entropy', max_depth=4, min_samples_split=9)  #

# 直接使用 GRID 的最优参数
tc = DecisionTreeRegressor(**grid.best_params_)

tc.fit(x, y)
tc.get_params
```

```python
# 画图

from IPython.display import Image  
from sklearn import tree
import pydotplus

dot_data = tree.export_graphviz(tc,                    #tc 是对应分类器
                         out_file=None,  
                         feature_names=x.columns,      #对应特征的名字
                         class_names=y,#.unique(),                #对应类别的名字
                         filled=True, rounded=True,  
                         special_characters=True,
                         impurity=False)  
graph = pydotplus.graph_from_dot_data(dot_data)  
graph.write_png('example.png')    #保存图像
Image(graph.create_png()) 
```
